```python
import pandas as pd
import numpy as np
import random

# Set seed for reproducibility
np.random.seed(0)

# Sample data
genders = ['Male', 'Female', 'Non-binary']
classes = ['Math', 'Science', 'English', 'History', 'Art', 'Physical Education', 'Music', 'Computer Science']
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']

# Generate data
data = {
    'id': range(1, 51),
    'gender': [random.choice(genders) for _ in range(50)],
    'teachable_classes': [', '.join(random.sample(classes, k=random.randint(1, 4))) for _ in range(50)],
    'availability': [', '.join(random.sample(days, k=random.randint(1, 5))) for _ in range(50)],
    'max_hours_per_day': [random.randint(1, 8) for _ in range(50)],
    'max_hours_per_week': [random.randint(10, 40) for _ in range(50)]
}

# Create DataFrame
teachers_df = pd.DataFrame(data)

# Display the first few rows of the DataFrame
teachers_df.head()
```

```python
# Set seed for reproducibility
np.random.seed(1)

# Sample data for students
names = ['John', 'Jane', 'Alex', 'Emily', 'Chris', 'Katie', 'Michael', 'Sarah', 'David', 'Laura', 'Daniel', 'Emma']
grades = [9, 10, 11, 12]
class_placements = ['Math', 'Science', 'English', 'History', 'Art', 'Physical Education', 'Music', 'Computer Science']
student_genders = ['Male', 'Female', 'Non-binary']

# Generate data
student_data = {
    'id': range(1, 51),
    'name': [random.choice(names) + " " + random.choice(['Smith', 'Johnson', 'Williams', 'Brown', 'Jones']) for _ in range(50)],
    'grade': [random.choice(grades) for _ in range(50)],
    'class_placement': [random.choice(class_placements) for _ in range(50)],
    'acceptable_teachers': [', '.join(map(str, random.sample(range(1, 51), k=random.randint(1, 5)))) for _ in range(50)],
    'gender': [random.choice(student_genders) for _ in range(50)],
    'availability': [', '.join(random.sample(days, k=random.randint(1, 5))) for _ in range(50)],
    'max_hours_per_day': [random.randint(1, 8) for _ in range(50)],
    'max_hours_per_week': [random.randint(10, 40) for _ in range(50)]
}

# Create DataFrame
students_df = pd.DataFrame(student_data)

# Display the first few rows of the DataFrame
students_df.head()
```
